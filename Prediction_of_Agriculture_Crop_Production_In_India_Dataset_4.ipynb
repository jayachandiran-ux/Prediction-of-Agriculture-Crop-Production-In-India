```python
average = [
    df['2004-05'].mean(),
    df['2005-06'].mean(),
    df['2006-07'].mean(),
    df['2007-08'].mean(),
    df['2008-09'].mean(),
    df['2009-10'].mean(),
    df['2010-11'].mean(),
    df['2011-12'].mean()
]

years = [
    '2004-05','2005-06','2006-07','2007-08',
    '2008-09','2009-10','2010-11','2011-12'
]

plt.figure(figsize=(8,5))
plt.plot(years, average, marker='o')
plt.title("Average Production Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Production")
plt.grid(True)
plt.show()
```

```python
average = {
    '2004-05': df['2004-05'].mean(),
    '2005-06': df['2005-06'].mean(),
    '2006-07': df['2006-07'].mean(),
    '2007-08': df['2007-08'].mean(),
    '2008-09': df['2008-09'].mean(),
    '2009-10': df['2009-10'].mean(),
    '2010-11': df['2010-11'].mean(),
    '2011-12': df['2011-12'].mean()
}

print("Highest Production Year:", max(average, key=average.get))
print("Lowest Production Year :", min(average, key=average.get))
```
