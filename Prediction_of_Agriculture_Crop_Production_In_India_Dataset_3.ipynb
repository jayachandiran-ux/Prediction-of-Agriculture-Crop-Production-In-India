```python
df = df.drop(columns=['Unnamed: 4'])
```

```python
df['Season/ duration in days'] = df['Season/ duration in days'].fillna(
    df['Season/ duration in days'].mode()[0]
)
```

```python
df.isnull().sum()
```

```python
df['Recommended Zone'] = df['Recommended Zone'].fillna(
    df['Recommended Zone'].mode()[0]
)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Crop', data=df)
plt.xticks(rotation=45)
plt.title("Crop Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='Variety', data=df)
plt.xticks(rotation=90)
plt.title("Variety Distribution")
plt.show()
```
