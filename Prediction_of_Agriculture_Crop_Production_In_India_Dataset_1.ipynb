```python
plt.figure(figsize=(8,5))
sns.countplot(x='Crop', data=df)
plt.xticks(rotation=45)
plt.title("Crop Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='State', data=df)
plt.xticks(rotation=45)
plt.title("State Distribution")
plt.show()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.rename(columns={
    'Cost of Cultivation (`/Hectare) A2+FL': 'Cultivation_A2FL',
    'Cost of Cultivation (`/Hectare) C2': 'Cultivation_C2',
    'Cost of Production (`/Quintal) C2': 'Production_Cost',
    'Yield (Quintal/ Hectare)': 'Yield'
}, inplace=True)
```

```python
df.columns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Yield"], bins=10, kde=True)
plt.title("Yield Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df["Yield"])
plt.title("Yield Box Plot")
plt.show()
```

```python
for col in df.columns:
    print(repr(col))
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Crop', y='Yield', data=df)
plt.xticks(rotation=45)
plt.title("Average Yield by Crop")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='State', y='Yield', data=df)
plt.xticks(rotation=45)
plt.title("Average Yield by State")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Cultivation_C2', y='Yield', data=df)
plt.title("Cultivation Cost vs Yield")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Production_Cost', y='Yield', data=df)
plt.title("Production Cost vs Yield")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
